```python
base_model = "mistralai/Mistral-7B-v0.1"
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    base_model,
    padding_side="right",
    add_eos_token=True,
    )

tokenizer.pad_token = tokenizer.eos_token
```

```python
tokenizer.add_bos_token, tokenizer.add_eos_token
```

```python
import torch
```

```python
from transformers import BitsAndBytesConfig
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=False,
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
from transformers import AutoModelForCausalLM
```

```python
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=bnb_config,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)
```

```python
from datasets import load_dataset
```

```python
dataset_name = "umarigan/turkiye_finance_qa"
train_dataset = load_dataset(dataset_name, split="train")
```

```python
def generate_promt(sample):
  full_promt = f"""<s>[INST]{sample["soru"]}
                      [/INST]{sample["cevap"]}</s>
                      """
  return {"text":full_promt}
```

```python
generate_promt(train_dataset[0])
```

```python
generetad_train_dataset = train_dataset.map(generate_promt, remove_columns=list(train_dataset.features))
```

```python
generetad_train_dataset
```

```python
from peft import prepare_model_for_kbit_training
```

```python
model.gradient_checkpointing_enable()
model = prepare_model_for_kbit_training(model)
```

```python
def print_trainable_parameters(model):
  trainable_params = 0
  all_param = 0
  for _, param in model.named_parameters():
    all_param += param.numel()
    if param.requires_grad:
      trainable_params += param.numel()
  print(f"trainable params: {trainable_params} || all params: {all_param} || ratio%: {100*trainable_params/all_param}")
```

```python
from peft import LoraConfig, get_peft_model
```

```python
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj","k_proj","v_proj","o_proj","gate_proj","up_proj","down_proj","lm_head"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)
```

```python
model = get_peft_model(model, lora_config)
```

```python
print_trainable_parameters(model)
```

```python
print(model)
```

```python
from transformers import TrainingArguments
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=1,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    save_strategy="steps",
    save_steps=25,
    learning_rate=2e-5,
    weight_decay=0.001,
    max_steps=20,
    evaluation_strategy="steps",
    eval_steps=25,
    do_eval = True,
    report_to="none"
    )
```

```python

training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=8,
    gradient_accumulation_steps=4,
    optim="paged_adamw_32bit",
    save_strategy="epoch",
    learning_rate=2e-5,
    weight_decay=0.01,
    evaluation_strategy="epoch",
    do_eval=True,
    report_to="none",
    load_best_model_at_end=True
)
```

```python
from trl import SFTTrainer
```

```python
trainer = SFTTrainer(
    model=model,
    tokenizer = tokenizer,
    args=training_args,
    train_dataset=generetad_train_dataset,
    eval_dataset=generetad_train_dataset,
    peft_config=lora_config,
    dataset_text_field="text"
)
```

```python
model.config.use_cache = False
```

```python
trainer.train()
```

```python
model_finetuned_finance = "mistral-7b-turkish-finance"
```

```python
trainer.model.push_to_hub(model_finetuned_finance)
```
